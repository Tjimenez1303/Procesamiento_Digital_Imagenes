```python
#Packages:
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as nd
from skimage import io, color
```

```python
img=io.imread("/content/yo.png") #Reading the image
img.shape
```

```python
plt.imshow(img) #Plotting the image
```

```python
#How in this case the color doesn´t have information, then we change to gray scale:
img_gray=color.rgb2gray(img)
img_gray.shape
```

```python
plt.imshow(img_gray)
```

```python
#Aplying a convolution to the image:
kernel=np.array([[1,0,-1],[1,0,-1],[1,0,-1]]) #Kernel for detect vertical edges
img_vert_edg=nd.convolve(img_gray,kernel) #Convolution between the image and the kernel
img_vert_edg.shape #Obviously to do the convolution always reduct the resolution of the image
```

```python
#How we can see, efectly, the convolution detected the vertical edges:
plt.subplot(1,2,1)
plt.imshow(img_gray,cmap="gray")
plt.title("Input Image")
plt.subplot(1,2,2)
plt.imshow(img_vert_edg,cmap="gray")
plt.title("Output Image")
```

```python
#Doing the same for detect horizontal edges:
kernel=np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
img_hori_edges=nd.convolve(img_gray,kernel)
img_hori_edges.shape
```

```python
#How we can see, efectly, the convolution detected the horizontal edges:
plt.subplot(1,2,1)
plt.imshow(img_gray,cmap="gray")
plt.title("Input Image")
plt.subplot(1,2,2)
plt.imshow(img_hori_edges,cmap="gray")
plt.title("Output Image")
```

```python
#Doing the same for detect get a black background:
kernel=np.zeros((3,3))
img_black_back=nd.convolve(img_gray,kernel)
img_black_back.shape
```

```python
#How we can see, efectly, the convolution result a black background:
plt.subplot(1,2,1)
plt.imshow(img_gray,cmap="gray")
plt.title("Input Image")
plt.subplot(1,2,2)
plt.imshow(img_black_back,cmap="gray")
plt.title("Output Image")
```
